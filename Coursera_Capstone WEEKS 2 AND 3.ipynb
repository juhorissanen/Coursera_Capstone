```python
# Let's import the basic libraries and get to to it. 

import pandas as pd
import numpy as np

# Firsly let's take a look how the data is loooking in general -> I had some problems opening the data so i decided to change the encoding to ISO-8859-1 and note the delimiter and use the "r" for raw version
df =pd.read_csv(r"http://avoindata.valvira.fi/alkoholi/alkoholilupa_toimipaikkatiedot_ABC.csv", encoding='ISO-8859-1', delimiter=';')

df

# Okay, we have almost 15.0000 places in finland which have the right to sell alcohol. Let's narrow it down a bit. 

```

```python
df.columns

#Ok, so the column names are in Finnish. In English they are translated as followed: 

#NIMI = NAME (place of the bar or restaurant)
#OSOITE = Address  
#POSTINUMERO = Postal Code
#POSTITOIMIPAIKKA = City 
#KUNTA = The Town (might differ from city)
#LUVANALKU PVM = The date when the permission for sale of alcohol was granted 
#LUPATYYPPI = The type of permission granted - Here we are interested in the “Serving A, all alcoholic beverages”-permission" Which is in finnish "Anniskelu A, kaikki alkoholijuomat"
#Y-Tunnus = The Finnish Business ID - Each company in finland has to have it to legally operate.  
```

```python
#Let's drop the blanks. 

df.dropna()
df.dropna(subset=['Y-TUNNUS'], inplace=True)

# Let's sort values alphabetically according the addrees. 

df_sort = df.sort_values(by=["OSOITE"], ascending=True)

#Dropping all the other values but the wanted Values from the city of Kuopio and narrowing down the results to “Serving A, all alcoholic beverages”-permission" which is in Finnish Anniskelu A, kaikki alkoholijuomat 
df1 = df[df_sort['POSTITOIMIPAIKKA'] == 'Kuopio']
df_kpo = df1[df_sort['LUPATYYPPI'] == 'Anniskelu A, kaikki alkoholijuomat']

#Noticed a blank at the column and corrected and reset the index. 
df_kpo.rename(columns={'NIMI ': 'NIMI'}, inplace=True)
df_kpo.set_index("NIMI", inplace=True)

df_kpo

#Now, it seems that we have 121 restaurants applying for the criteria in the city Of KUOPIO. 
```

```python
# In order to better vizualize the the results I will be using geopy and folium. Geopy helps when I am trying to get the GPS cordinates 
# and folium helps me to map the places as we get there. 

# I will test the geopy by inserting it to locate the City of Kuopio to see whether its working. 

!conda install -c conda-forge geopy --yes 
from geopy.geocoders import Nominatim # module to convert an address into latitude and longitude values
locator = Nominatim(user_agent="abc")
location = locator.geocode("Kuopio")
!conda install -c conda-forge folium=0.5.0 --yes
import folium # plotting library
print(location.address)
print((location.latitude, location.longitude))
```

```python
# Seems that Geopy and Folium have been installed just fine. Let's take a look at the dataframe again. 

df_kpo
```

```python
# OK. All good. Now I need to combine the Address and City columns (OSOITE and POSTITOIMIPAIKKA) to get the geopy working correctly, since the same adrees can be the same in multiple cities in finland.
# I solved this just by adding a blank colum and combining these three columns
# After this I created a new column called "LOKAATIO" which means "LOCATION" in english. This is the information wich will work with geopy to proivde the cordiantes. 

df_kpo[','] = ", "
df_kpo["LOKAATIO"] = df_kpo["OSOITE"] + df_kpo[","] + df_kpo["POSTITOIMIPAIKKA"]
df_kpo.drop(columns=",",inplace=True)
df_kpo["Sijainti"] = df_kpo["LOKAATIO"].apply(locator.geocode)

# Let's have look at columns now. The "Sijainti" Column refers to location and it is a result of using geopy for the created "Location Columns"

df_kpo
```

```python
# Now it is time to get the cordinates out from the location column created with geopy. Let's call cordiantes to columns 'point'
df_kpo['point'] = df_kpo['Sijainti'].apply(lambda loc: tuple(loc.point) if loc else None)
```

```python
df_kpo

# And there they are! (Last column on the right "Point")
```

```python
#Some checks
df_kpo['point'][0][0]
```

```python
#Now I will be running the cordintes to different columns in order that they will be available for mapping. 

df_kpo[['latitude', 'longitude', 'altitude']] = pd.DataFrame(df_kpo['point'].tolist(), index=df_kpo.index)
```

```python
# Let's drop the columns which are not needed. 
df_kpo = df_kpo.drop(['POSTINUMERO', 'OSOITE', 'KUNTA','LUVANALKUPVM', 'Sijainti', 'point'], axis=1)
```

```python
# And let's drop the columns with blanks. Just in case if the geopy did not find matching addresses for all. 
df_kpo1 = df_kpo.dropna()
```

```python
# Now, let's have look at our final data. It looks good. 
df_kpo1

# It looks like we are left wiht 100 restaurants with the wanted criteria and with the needed location data to be mapped. 
```

```python
# Just making sure that we have the latitude parameters without nulls. 

df_kpo1.latitude.isnull().sum()
```

```python
# Now, it is time to map the results. 

#1 Let's create a basic map, I took the staring coordinates from the results list. 
map1 = folium.Map(
    location=[62.89224,27.690307],
    tiles='CartoDB positron',
    zoom_start=12,)
map1
```

```python
# Now, as we can see that the map is working, let's make our first try to map the places where to get beer in "KUOPIO"
df_kpo1.apply(lambda row:folium.CircleMarker(location=[row["latitude"], row["longitude"]]).add_to(map1), axis=1)
map1
```

```python
# This was good, but I think we can get even a better map. 
```

```python
# Let's use Market cluster 

from folium.plugins import MarkerCluster
from folium.plugins import FastMarkerCluster
folium_map = folium.Map(location=[62.89224,27.690307],
                        zoom_start=12,
                        tiles='CartoDB positron')


FastMarkerCluster(data=list(zip(df_kpo1['latitude'].values, df_kpo1['longitude'].values))).add_to(folium_map)
folium.LayerControl().add_to(folium_map)

folium_map

#Looking good 
```

```python
# I had a problem to get the names displayed on the map so I decided to do it another way. 
```

```python
# let's get the locations and chech that we have 100 of them. 
locations = df_kpo1[['latitude', 'longitude']]
locationlist = locations.values.tolist()
len(locationlist)
```

```python
# It is a good idea to rename the index get it to good use. 
df_kpo1["NIMI"] = df_kpo1.index
```

```python
#Finally, let's map the places with names so we get the information that I was asking in the beginning. 

beer_map = folium.Map(location=[62.89224,27.690307],
                        zoom_start=12,
                        tiles='CartoDB positron')


for point in range(0, len(locationlist)):
    folium.Marker(locationlist[point], popup=df_kpo1["NIMI"][point]).add_to(beer_map)



beer_map

# Now, we can answered questions wich I wanted to get answers to:  
# A. What are the places holding “Serving A, all alcoholic beverages”-permission for their business, meaning that they can serve all kind of alcoholic drinks, including beer.
# Here they are, all mapped. Answering the question B-part "b.	Where are these places located in the city of of my interest, Kuopio? "
```
